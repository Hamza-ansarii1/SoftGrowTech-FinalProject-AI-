```python
#import libraries
import nltk
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#download tokenizer
nltk.download('punkt')
```

```python
#question & response
data = {
    "hello":"Hello How can i help you?",
    "hi":"hi Welcome.",
    "price":"please check our website for prices",
    "refund":"Refund are available within 15 days.",
    "delivery":"Delivery takes max 5 days.",
    "payment":"We accept Visa,MasterCard,Nayapay, & JazzCash",
    "contact":"Email us at support@gmail.com",
    "thanks":"You're welcome",
    "bye":"Goodbye have a nice day"
}

questions = list(data.keys())
answers = list(data.values())
```

```python
# TF-IDF model
vectorizer = TfidfVectorizer()
question_vectors = vectorizer.fit_transform(questions)

print("AI Costomer Support Chatbot")
print("Type 'bye' to exit.\n")

while True:
    user_input = input("You: ").lower()

    if user_input == "bye":
        print("Bot:",data["bye"])
        break
        
   #convert input into vector
    user_vector = vectorizer.transform([user_input])

    
 #calculate similarity
    similarity = cosine_similarity(user_vector,question_vectors)

    best_match = similarity.argmax()
    score = similarity[0][best_match]

    #threshold for response
    if score > 0.3:
         print("Bot:",answers[best_match])
    else:
         print("Bot: Sorry, I couldn't understand")
               
```
